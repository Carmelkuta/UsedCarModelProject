```python
import numpy as np # used for scientific computing
import pandas as pd # used for data analysis and manipulation
import matplotlib.pyplot as plt # used for visualization and plotting
from sklearn.impute import SimpleImputer # Used to fit in missing data
from sklearn import preprocessing # used for label encoder
from sklearn.experimental import enable_iterative_imputer #imputations for NaN values
from sklearn.impute import IterativeImputer #Imputations for NaN values
```

```python
df = pd.read_csv('vehicles.csv')
```

```python
df = df.drop(labels=range(0,27), axis=0)
df.head(30)
```

```python
def preprocess(X, y):
    """Preforming a mean normalization to the data which will help speed up the process
    Inputs:
    X - Inputs (n features over m instances)
    y - True Labels
    
    Output:
    X - Mean normalized inputs
    y - Mean normalized true labels"""
    
    for i in range(len(X)):
        stdX = X[i].std()
        meanX = X[i].mean()
        
        for j in range(len(X[i])):
            X[i][j] = (X[i][j] - meanX) / stdX
    
    for i in range(len(y)):
        y[i] = (y[i] - y.mean()) / y.std()
        
    return X, y
```

```python
### Getting Rid of Outliers ###

betterdf = df[df['odometer'] < 150000]
betterdf = betterdf[betterdf['odometer'] > 50]
betterdf = betterdf[betterdf['year'] > 1940]
betterdf = betterdf[betterdf['price'] > 500]
betterdf = betterdf[betterdf['price'] < 200000]
print(len(betterdf))
print(len(betterdf))
X = betterdf['year'].values
y = betterdf['price'].values
X = X.astype(float)
X = X.reshape(1,302048)
print(X.shape)
```

```python
# training and validation split
np.random.seed(42)
indices = np.random.permutation(X.shape[1])
idx_train, idx_val = indices[:int(0.8*X.shape[1])], indices[int(0.8*X.shape[1]):]

X_train, X_val = X[0][idx_train], X[0][idx_val]
y_train, y_val = y[idx_train], y[idx_val]


X_val = X_val.reshape(1, 60410)
print(X_val.shape)
X_val, y_val = preprocess(X_val, y_val)

X_val = X_val.reshape(60410,)
```

```python
def labelEncoder(X):
    le = preprocessing.LabelEncoder()
    for i in range(len(X)):
        if (i != 1) and (i != 7) and (i != 13) and (i != 14):
            for j in range(len(X[i])):
                if (i != 1) and (i != 7) and (i != 13) and (i != 14):
                    if X[i][j] != X[i][j]:   #NaN is considered a float and we convert to string
                        X[i][j] = '-1'
                    
            le.fit(X[i])
            X[i] = le.transform(X[i])
    return
```
